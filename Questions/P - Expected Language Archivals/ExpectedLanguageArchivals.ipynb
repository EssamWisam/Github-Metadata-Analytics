</span>

#### <font color="LemonChiffon">Monthly Time Series Forecasting</font> with XGBoost over the Fraction of Archivals of each Language

We will consider all languages that were used in 2021 with at least 50 repos and 3 archivals. Call this out suspected list.

```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from Logic import train_test_timesplit, plot_timeperformance, plot_future_2023
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from tqdm import tqdm

susp_list = get_suspect_langs(x_data_t)
lang_preds = { lang: {"train_pred":None, "val_pred":None, "rmse":None } for lang in susp_list }

for lang in tqdm(susp_list):
    x_data_lang = archival_overtime_given_lang(x_data_d, lang, unit='month')

    x_train, y_train, x_val, y_val = train_test_timesplit(x_data_lang, split_date='2021-01-01')

    # fit XGBoost model to training data
    model = XGBRegressor(booster='gbtree', max_depth = 5, gamma = 0.00015)
    model.fit(x_train, y_train)

    # make predictions on training and valing data
    y_train_preds = model.predict(x_train)
    y_val_preds = model.predict(x_val)

    # calculate RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_val_preds))

    lang_preds[lang]["x_train"], lang_preds[lang]["y_train"]  = x_train, y_train
    lang_preds[lang]["x_val"], lang_preds[lang]["y_val"] = x_val, y_val
    lang_preds[lang]["train_pred"], lang_preds[lang]["val_pred"] = y_train_preds, y_val_preds
    lang_preds[lang]["rmse"] = rmse
```

#### Now Train on all Data then Predict for 2023 and Average the Predictions

```python
lang_archives_2023 = { lang: None for lang in susp_list }

for lang in tqdm(susp_list):
    # Now let's train model on the entire data and make predictions on the next 12 months
    x_data_lang = archival_overtime_given_lang(x_data_d, lang, unit='month')
    x_train, y_train, _, _ = train_test_timesplit(x_data_lang, split_date=None)
    model.fit(x_train, y_train)
    y_train_preds = model.predict(x_train)
    train_rmse = np.sqrt(mean_squared_error(y_train, y_train_preds))

    # Let's predict the next 12 months
    x_test = pd.DataFrame(pd.date_range(start='2022-12-01', end='2023-12-01', freq='MS'), columns=['year_month'])
    x_test['year'] = x_test['year_month'].dt.year
    x_test['month'] = x_test['year_month'].dt.month
    x_test = x_test.set_index('year_month')

    # make predictions on test data
    y_test_preds = model.predict(x_test)
    y_test_preds[y_test_preds < 0] = 0

    avg_archives = np.mean(y_test_preds)
    lang_archives_2023[lang] = avg_archives
```

```python
plot_future_2023(lang_archives_2023)
```
