### <font color="LemonChiffon"> Hypothesis regarding the Low Association of Languages via Stars </font>

Lets test the following Hypothesis:
<br><br>
$H_0$: It doesn't hold that $16<μ^{l}_{stars}<19$ for one or more languages $l$.
<br><br>
$H_1$: It holds that $16<μ^{l}_{stars}<19$ for each language $l$.

We will set the significance level $\alpha=0.05$ and perform two one-sided tests for each language.

```python
import scipy.stats as stats
from Logic import hyp_test_result, normality_test

μs = [15.5, 19.5]
lang_results_stars = { lang: [] for lang in top_langs}
for μ in μs:
    for lang in top_langs:
        t, p = stats.ttest_1samp(splits[lang]["stars"], μ)
        lang_results_stars[lang].append(hyp_test_result("μ", str(μ), p, t, 0.05, one_sided=True))

display(HTML(nice_table(lang_results_stars, "One-sided t-test for Mean of Stars Results")))
```

### <font color="LemonChiffon">Hypothesis regarding the High Commit Activity in TypeScript compared to Other Languages <font>

Lets test the following Hypothesis where $TS$ indicates the TypeScript language:
<br><br>
$H_0$: It holds that $μ^{TS}_{Commits}\leqμ^{l}_{Commits}$ for one or more language $l$.
<br><br>
$H_1$: It holds that $μ^{TS}_{Commits}>μ^{l}_{Commits}$ for each language $l$.

We will set the significance level $\alpha=0.05$ and perform a one-sided tests for each language.

```python

lang_results_commits = { lang: [] for lang in top_langs if lang != "TypeScript"}
for lang in top_langs:
    if lang == "TypeScript": continue
    t, p = stats.ttest_ind(splits["TypeScript"]["defaultBranchCommitCount"], splits[lang]["defaultBranchCommitCount"])
    lang_results_commits[lang].append(hyp_test_result(f"μ(TypeScript)", f"μ({lang})", p, t, 0.05, one_sided=True))

display(HTML(nice_table(lang_results_commits, "One-sided two-sample t-test for Mean of Commits Results")))
```

### <font color="LemonChiffon"> Hypothesis regarding the Poor Community Engagement in C compared to Other Languages </font>

Lets test the following Hypothesis where $C$ indicates the C language:
<br><br>
$H_0$: It holds that $μ^{C}_{PullRequests}\geqμ^{l}_{PullRequests}$ for one or more language $l$.
<br><br>
$H_1$: It holds that $μ^{C}_{PullRequests}<μ^{l}_{PullRequests}$ for each language $l$.

We will set the significance level $\alpha=0.05$ and perform a one-sided tests for each language.

```python
lang_results_pulls = { lang: [] for lang in top_langs if lang != "C"}
for lang in top_langs:
    if lang == "C": continue
    t, p = stats.ttest_ind(splits["C"]["pullRequests"], splits[lang]["pullRequests"])
    lang_results_pulls[lang].append(hyp_test_result(f"μ(C)", f"μ({lang})", p, t, 0.05, one_sided=True))

display(HTML(nice_table(lang_results_pulls, "One-sided two-sample t-test for Mean of Pull Requests Results")))
```
