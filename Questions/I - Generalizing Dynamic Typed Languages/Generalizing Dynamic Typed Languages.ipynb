<font color="yellow"> <h2> Hypothesis </font> <font color='pink'> Testing </font></h2> 
</span>

<span style="font-family:;">
<font color="Yellow"> <h3>Hypothesis: </font> The average usage of <font color='pink'>dynamically typed </font>programming languages is greater than the average usage of <font color='pink'>statically typed</font> ones</h3>
</span>

## T-Test

```python
Logic = importlib.reload(Logic)
dynamic_datasets, static_datasets = Logic.split_dataset_across_years_months(ds)


dynamic_usage = [(year, month, len(dynamic_datasets[year][month])) for year in dynamic_datasets.keys() for month in dynamic_datasets[year].keys()]
static_usage = [(year, month, len(static_datasets[year][month])) for year in static_datasets.keys() for month in static_datasets[year].keys()]

dynamic_usage.sort(key=lambda x: (x[0], x[1]))
static_usage.sort(key=lambda x: (x[0], x[1]))

dynamic_usage = [x[2] for x in dynamic_usage]
static_usage = [x[2] for x in static_usage]

#Perform independent samples t-test
t_stat, p_value = stats.ttest_ind(dynamic_usage, static_usage, alternative='greater')

alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis (H0) in favor of the alternative hypothesis (Ha)")
    print("Average usage of dynamic programming languages is significantly greater than the average usage of static programming languages.")
else:
    print("Fail to reject the null hypothesis (H0)")
    print("Average usage of dynamic programming languages is not significantly greater than the average usage of static programming languages.")
```

```python
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

## Z-Test

```python
dynamic_mean = np.mean(dynamic_usage)
static_mean = np.mean(static_usage)
dynamic_std = np.std(dynamic_usage)
static_std = np.std(static_usage)

# Calculate the standard error of the difference between the means
n1 = len(dynamic_usage)
n2 = len(static_usage)
standard_error = np.sqrt((dynamic_std**2 / n1) + (static_std**2 / n2))

# Perform z-test
z_stat = (dynamic_mean - static_mean) / standard_error
p_value = 1 - stats.norm.cdf(z_stat)

alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis (H0) in favor of the alternative hypothesis (Ha)")
    print("Average usage of dynamic programming languages is significantly greater than the average usage of static programming languages.")
else:
    print("Fail to reject the null hypothesis (H0)")
    print("Average usage of dynamic programming languages is not significantly greater than the average usage of static programming languages.")
```

```python
print("z-statistic:", z_stat)
print("p-value:", p_value)
```

## Proporition Test

```python
import statsmodels.api as sm

dynamic_total = sum(dynamic_usage)
static_total = sum(static_usage)

dynamic_proportion = dynamic_total / (dynamic_total + static_total)
static_proportion = static_total / (dynamic_total + static_total)

# Perform proportion test
successes = [dynamic_total, static_total]
totals = [dynamic_total + static_total, dynamic_total + static_total]
z_stat, p_value = sm.stats.proportions_ztest(successes, totals, alternative='larger')

alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis (H0) in favor of the alternative hypothesis (Ha)")
    print("Dynamic programming languages are more popular than static programming languages.")
else:
    print("Fail to reject the null hypothesis (H0)")
    print("There is insufficient evidence to suggest that dynamic programming languages are more popular than static programming languages.")
```
